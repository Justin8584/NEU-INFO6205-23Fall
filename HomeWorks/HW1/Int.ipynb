# class Int

```python
#WRITE class Int
#Author: Jiading Zhou, 2023 Fall Semester, NEU University
############################################################

class Int:
    def __init__(self, num=0):
        self.sign = -1 if num < 0 else 1
        self.digits = [int(digit) for digit in str(abs(num))]

    def __str__(self):
        return f"{'-' if self.sign == -1 else ''}{''.join(map(str, self.digits))}"

    def int(self):
        return self.sign * int(''.join(map(str, self.digits)))

    def __len__(self):
        return len(self.digits)

    def __getitem__(self, index):
        return self.digits[index]

    def __setitem__(self, index, value):
        self.digits[index] = value

    def __add__(self, other):
        return Int(self.int() + other.int())

    def __sub__(self, other):
        return Int(self.int() - other.int())

    def __mul__(self, other):
        return Int(self.int() * other.int())

    def __lt__(self, other):
        return 1 if self.int() < other.int() else 0

    def __le__(self, other):
        return 1 if self.int() <= other.int() else 0

    def __eq__(self, other):
        return 1 if self.int() == other.int() else 0

    def __ne__(self, other):
        return 1 if self.int() != other.int() else 0

    def __gt__(self, other):
        return 1 if self.int() > other.int() else 0

    def __ge__(self, other):
        return 1 if self.int() >= other.int() else 0
```
