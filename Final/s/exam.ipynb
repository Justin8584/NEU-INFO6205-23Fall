```python
class Write2Dot:
    def __init__(self, d: "bool", start: "int", a: "list of list", ans: "list of list",f: "filename",) -> "None":
        
        print("out graph", f)
#         print("ZERO marks will be given if you don't write code here")

        self.graph = "graph"
        self.a = a
        self.ans = ans
        self.f = f

        self.write_dot()

    def write_dot(self):
        dot_content = "## Jiading Zhou ####\n"
        dot_content += "graph g {\n"
        dot_content += "    overlap=false; splines=true;\n"
        dot_content += "    edge [style=dotted, weight=10, len=.2];\n"
        
        if self.ans:
            boys, girls = self.ans
            dot_content += "    subgraph cluster_RED {\n"
            dot_content += "        RED [pos=\"-1,0!\", color=red];\n"
            for node in boys:
                dot_content += f"        {node} -- RED;\n"
            dot_content += "    }\n"
            
            dot_content += "    subgraph cluster_BLUE {\n"
            dot_content += "        BLUE [pos=\"-1,0!\", color=blue];\n"
            for node in girls:
                dot_content += f"        {node} -- BLUE;\n"
            dot_content += "    }\n"

        dot_content += "    edge [style=\"\", weight=1, len=1];\n"
        for i, nodes in enumerate(self.a):
            for j in nodes:
                if i < j:  
                    dot_content += f"    {i} -- {j};\n"

        # Convert list of integers to a comma-separated string
        boys_str = ', '.join(map(str, boys))
        girls_str = ', '.join(map(str, girls))

        # Add the label showing the bipartite sets without additional square brackets
        dot_content += f"    label= \"[{boys_str}][{girls_str}]\";\n"
        dot_content += "}\n"

        with open(self.f, 'w') as file:
            file.write(dot_content)
```

```python
############################################################
# Exam.py 
# Jiading Zhou - INFO 6205 Final
###########################################################

############################################################
#  All imports here
###########################################################
class Exam:
    def __init__(
        self,
        num: "int",
        d: "bool",
        start: "int",
        a: "list of list",
        ans: "list of list",
        work: "list of size 1",
        show: "bool",
    ):
        self._num = num
        self._dir = d
        self._start = start
        self._a = a
        self._ans = ans
        self._work = work
        self._show = show
        ## You can have your data structure here
        
        self._colors = [-1] * len(a) if a else []
        

        ## Nothing can be changed below
        if self._show:
            # CHANGE THIS. MUST post all pdf in Canavs as a ZIP file
            # DO NOT UPLOAD DOT FILES
            outputDir = "/Users/jiadingzhou/Desktop/info6205/Final/s/"
            f = outputDir + str(self._num) + "in.dot"
            e = WriteDot(self._dir, self._start, self._a, f)
            
        self._alg()  # Everything happens in _alg
        if self._show:
            f = outputDir + str(self._num) + "out.dot"
            e = Write2Dot(self._dir, self._start, self._a, self._ans, f)

    ############################################################
    #  Write your code below
    ###########################################################

    def _alg(self):
        if not self._a:
            return []
        
        for node in range(len(self._a)):
            if self._colors[node] == -1:
                if not self._dfs(node, 0):
                    self._ans.append([])
                    return self._ans  

        boys, girls = [], []
        for node, color in enumerate(self._colors):
            if color == 0:
                boys.append(node)
            elif color == 1:
                girls.append(node)
        self._ans.append(boys)
        self._ans.append(girls)
        return self._ans  

    def _dfs(self, node, color):
        if self._colors[node] != -1:
            return self._colors[node] == color

        self._colors[node] = color
        for neighbor in self._a[node]:
            if not self._dfs(neighbor, 1 - color):
                return False

        return True
```
